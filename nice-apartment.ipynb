```python
from bs4 import BeautifulSoup
import requests

def extract_apartment_info(url, info):
    r = requests.get(url)
    headers = {"X-Requested-With":"XMLHttpRequest", "User-Agent":"Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1985.125 Safari/537.36"}
    r = requests.get(url, headers=headers)
    data = r.text
    soup = BeautifulSoup(data, 'html.parser')
    apartments = soup.find_all("div", id=lambda value: value and value.startswith("annonce-"))
    
    for apartment in apartments:
        d = dict()
        d["description"] = apartment.findAll("div", { "class": "text g-col u-pad g-666 g-100-sm" })[0].text.strip()
        d["price"] = apartment.findAll("span", { "class": "c-pa-cprice" })[0].text.strip()
        d["criterion"] = apartment.findAll("div", { "class": "c-pa-criterion" })[0].text.strip()
        info.append(d)
    
```

```python
info = []
for i in range(5):
    url = "http://www.seloger.com/immobilier/locations/immo-nice-06/bien-appartement/?LISTING-LISTpg=" + str(i+1)
    extract_apartment_info(url, info)
```
